```python
from tkinter import *
from tkinter import messagebox as mb


window = Tk()
window.title("Quiz")
window.geometry("800x450")
window.config(background='white')


class Quiz:
    def __init__(self):
        self.q_no = 0
        self.opt_selected = IntVar()
        self.opts = self.radio_buttons()
        self.display_question()
        self.display_options()
        self.buttons()
        self.data_size = len(question_bank)
        self.correct = 0

    def display_question(self):
        q_no = Label(window, text=question_bank[self.q_no]["text"], width=60,
                     font=('ariel', 16, 'bold'), anchor='w', bg='white')
        q_no.place(x=70, y=100)

    def display_options(self):
        val = 0
        self.opt_selected.set(0)
        for option in options[self.q_no]:
            self.opts[val]['text'] = option
            val += 1

    def radio_buttons(self):
        q_list = []
        y_pos = 150
        while len(q_list) < 4:
            radio_btn = Radiobutton(window, text=" ", variable=self.opt_selected,
                                    value=len(q_list) + 1, font=("ariel", 14), bg='white')
            q_list.append(radio_btn)
            radio_btn.place(x=100, y=y_pos)
            y_pos += 40
        return q_list

    def buttons(self):
        next_button = Button(window, text="Next", command=self.next_btn,
                             width=10, bg="blue", fg="white", font=("ariel", 16, "bold"))
        next_button.place(x=350, y=380)

        quit_button = Button(window, text="Quit", command=window.destroy,
                             width=5, bg="black", fg="white", font=("ariel", 16, "bold"))
        quit_button.place(x=700, y=50)

    def check_ans(self, q_no):
        correct_answer = answer[q_no]
        return self.opt_selected.get() == correct_answer

    def next_btn(self):
        if self.check_ans(self.q_no):
            self.correct += 1
        
        self.q_no += 1
        
        if self.q_no == self.data_size:
            self.display_result()
            window.destroy()
        else:
            self.display_question()
            self.display_options()

    def display_result(self):
        wrong_count = self.data_size - self.correct
        correct = f"Correct: {self.correct}"
        wrong = f"Wrong: {wrong_count}"
        
        score = int(self.correct / self.data_size * 100)
        result = f"Score: {score}%"
        
        mb.showinfo("Result", f"{result}\n{correct}\n{wrong}")

question_bank = [
    {"text": "Who wrote the play Romeo and Juliet?"},
    {"text": "What is the chemical symbol for water?"},
    {"text": "Which planet is known as the Red Planet?"},
    {"text": "Who painted the Mona Lisa?"},
    {"text": "Which Indian city is known as the Silicon Valley of India?"}
]

options = [
    ["A: William Shakespeare", "B: Mark Twain", "C: George Orwell", "D: Charles Dickens"],
    ["A: O2", "B: CO2", "C: HO2", "D: H2O"],
    ["A: Earth", "B: Venus", "C: Mars", "D: Jupiter"],
    ["A: Vincent van Gogh", "B: Pablo Picasso", "C: Leonardo da Vinci", "D: Michelangelo"],
    ["A: Hyderabad", "B: Bangalore", "C: Chennai", "D: Pune"]
]

answer = [1, 4, 3, 3, 2]

quiz = Quiz()

window.mainloop()
```
